```python
print df.columns
df = df.dropna()
workclass = list(set(df.workclass.values))
education = list(set(df.education.values))
marital_status = list(set(df["marital-status"]))
occupation = list(set(df.occupation.values))
relationship = list(set(df["relationship "]))
race = list(set(df.race.values))
sex = list(set(df.sex.values))
country = list(set(df["native-country"]))
income = list(set(df.income.values))
print income

data = df.copy()

for i in range(len(workclass)):
    data.workclass[data.workclass == workclass[i]] = i

for i in range(len(education)):
    data.education[data.education == education[i]] = i

for i in range(len(marital_status)):
    data["marital-status"][data["marital-status"] == marital_status[i]] = i

for i in range(len(occupation)):
    data.occupation[data.occupation == occupation[i]] = i

for i in range(len(relationship)):
    data["relationship "][data["relationship "] == relationship[i]] = i
    
for i in range(len(race)):
    data.race[data.race == race[i]] = i
    
for i in range(len(sex)):
    data.sex[data.sex == sex[i]] = i
    
for i in range(len(country)):
    data["native-country"][data["native-country"] == country[i]] = i

for i in range(len(income)):
    data.income[data.income == income[i]] = i
```

# Pythonの機械学習ライブラリ

[scikit-learn](http://scikit-learn.org/stable/)

なんでもある。  
データだけ用意すれば、実装がすぐできる。  



```python
from sklearn.ensemble import RandomForestClassifier

train_data = data[0:int(data.shape[0]*0.8)]
test_data = data[0:int(data.shape[0]*0.8)]

train = train_data.ix[ : , 0:13]
label = train_data.ix[ : , -1]

estimator = RandomForestClassifier()

# 学習させる
estimator.fit(train, label)
```

```python
# 推定
test = test_data.ix[ : , 0:13]
label_predict = estimator.predict(test)
```

```python
# 精度評価
est_df = test_data.copy()
est_df["results"] = label_predict
accuracy = float(est_df.query('income == results').shape[0]) / float(est_df.shape[0])
print "Estimate accuracy: %0.2f %% (%d / %d)" % (accuracy * 100, est_df.query('income == results').shape[0], est_df.shape[0])
```

## Regression

### Forest Fires Data Set  
This is a difficult regression task, where the aim is to predict the burned area of forest fires, in the northeast region of Portugal, by using meteorological and other data.  
![](http://archive.ics.uci.edu/ml/assets/MLimages/Large162.jpg)
http://archive.ics.uci.edu/ml/datasets/Forest+Fires


### Attribute Information:

For more information, read [Cortez and Morais, 2007]. 
1. X - x-axis spatial coordinate within the Montesinho park map: 1 to 9 
2. Y - y-axis spatial coordinate within the Montesinho park map: 2 to 9 
3. month - month of the year: 'jan' to 'dec' 
4. day - day of the week: 'mon' to 'sun' 
5. FFMC - FFMC index from the FWI system: 18.7 to 96.20 
6. DMC - DMC index from the FWI system: 1.1 to 291.3 
7. DC - DC index from the FWI system: 7.9 to 860.6 
8. ISI - ISI index from the FWI system: 0.0 to 56.10 
9. temp - temperature in Celsius degrees: 2.2 to 33.30 
10. RH - relative humidity in %: 15.0 to 100 
11. wind - wind speed in km/h: 0.40 to 9.40 
12. rain - outside rain in mm/m2 : 0.0 to 6.4 
13. area - the burned area of the forest (in ha): 0.00 to 1090.84 
(this output variable is very skewed towards 0.0, thus it may make 
sense to model with the logarithm transform).

```python
df = pd.read_csv("forestfires.csv", sep=',')
```

```python
df
```

## ピボットテーブル（クロス集計）  
質問項目をかけ合わせて集計する手法。  
質問項目を1つの表の表頭と表側に分け、それぞれのカテゴリーが交わるセルに、該当する回答数や回答比率を記載する

```python
pd.pivot_table(df, values='area', index=['X', 'Y'], columns=['month'])
```

```python
month = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
day = {'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5, 'sat': 6, 'sun': 7}

data = df.copy()
for  m in month:
    data.month[data.month == m] = month[m]

for d in day:
        data.day[data.day == d] = day[d]
```

```python
data
```

```python
data.corr()
```

```python
# 学習データセット
train_data = data[0:int(data.shape[0] * 0.8)]
test_data = data[int(data.shape[0] * 0.8):]

train = train_data.ix[ : , 2:11]
label = train_data.ix[: , -1]

from sklearn import linear_model
clf = linear_model.LinearRegression()
est = clf.fit (train, label)
```

```python
clf.coef_
```

```python
test = test_data.ix[ : , 2:11]
test_label = test_data.ix[:, -1]
results = est.predict(test)
test_data["results"] = results
```
